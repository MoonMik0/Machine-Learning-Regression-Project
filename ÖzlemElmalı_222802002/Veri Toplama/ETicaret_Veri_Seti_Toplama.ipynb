### Veri Toplama

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
#gerekli kütüphaneler
```

```python
sayfa=1#site içinde sayfalar arası gezmek için 35'e kadar artacak bir değer
urunler_liste=[]#ürünün tüm bilgilerini sırasıyla tutacak liste
sutunlar = ['Fiyat', 'Ürün Adı', 'Kullanım Amaç', 'İşletim Sistemi', 'Ekran Boyutu', 'Ekran Çözünürlüğü', 'Ekran Hızı', 'İşlemci Markası', 'İşlemci Serisi', 'RAM', 'SSD Var mı', 'SSD Kapasitesi', 'GPU Modeli','GPU Var mı', 'USB 3.0 Var mı', 'USB 3.0 Adet', 'USB-C Var mı', 'USB-C Adet', 'HDMI Var mı', 'Pil Gücü', 'Pil Hücre Sayısı']
#sütun adları
veri_listesi=[]#sırasıyla tüm ürünlerin biliglerini tutacak liste
```

```python
while sayfa <=35:
    url = 'https://www.epey.com/laptop/e/YToxOntzOjQ6Im96ZWwiO2E6MTp7aTowO3M6Nzoic2F0aXN0YSI7fX1fTjs=/'#Veri çekmek için epey sitesi kullanıldı
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(url+str(sayfa), headers=headers)#Laptop biligleri çekilecek
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
    else:
        print('robots.txt dosyasına erişilemedi. Durum Kodu:', response.status_code)
    
    ul = soup.find_all("ul",attrs = {"class":"metin row"})
    if ul: #her sayfanın başında bir tane başka siteye giden reklam var onlar atıldı
        ul = ul[1:]
    for urun in ul :
        urun_linkleri = urun.find_all("li",attrs={"class":"adi cell"})#sayfadaki ürünlerin linkleri toplandı
   
        for i in urun_linkleri : 
            link_div = i.find("div",attrs={"class":"detay cell"})#her ürünün linki tek tek çağırıldı
            link = link_div.a.get("href")
            detay = requests.get(link, headers=headers)
            detay_soup = BeautifulSoup(detay.text, 'html.parser')

            #maalesef her ürünün detaylar kısmında her bilgi yok bazen bulunamadığı için null döndürüyor ama değişkenlere atanamıyor o yüzden her birinin null kontrolünün yapılması gerekiyor 
            urun_fiyat_element = detay_soup.find("div", attrs={"data-jplist-group": "group"}).find("a")
            if urun_fiyat_element:
                span_urun_fiyat = urun_fiyat_element.find("span", attrs={"class": "urun_fiyat"})
                if span_urun_fiyat:
                    urun_fiyat = span_urun_fiyat.text.strip()
                    urun_fiyat = urun_fiyat.split(" TL")[0]#TL kısmı çıktı sadece sayı kısmı kaldı
                else:
                    urun_fiyat = None
            else:
                urun_fiyat = None

            urun_adi_element = detay_soup.find("div", attrs={"class": "baslik"})
            if urun_adi_element:
                h1_urun_adi = urun_adi_element.find("h1")
                if h1_urun_adi:
                    urun_adi = h1_urun_adi.text.strip()
                else:
                    urun_adi = None
            else:
                urun_adi = None
                
            urun_amaci_element = detay_soup.find("li", attrs={"id": "id1493"})
            if urun_amaci_element:
                a_urun_amaci = urun_amaci_element.find("a")
                if a_urun_amaci:
                    urun_amaci = a_urun_amaci.text.strip()
                else:
                    urun_amaci = None
            else:
                urun_amaci = None

            isletim_sis_element = detay_soup.find("li", attrs={"id": "id1038"})
            if isletim_sis_element:
                a_isletim_sis = isletim_sis_element.find("a")
                if a_isletim_sis:
                    isletim_sis = a_isletim_sis.text.strip()
                else:
                    isletim_sis = None
            else:
                isletim_sis = None

            ekrn_byt_element = detay_soup.find("li", attrs={"id": "id1426"})
            if ekrn_byt_element:
                a_ekrn_byt = ekrn_byt_element.find("a")
                if a_ekrn_byt:
                    ekrn_byt = a_ekrn_byt.text.strip()
                else:
                    ekrn_byt = None
            else:
                ekrn_byt = None

            islemci_marka_element = detay_soup.find("li", attrs={"id": "id1367"})
            if islemci_marka_element:
                a_islemci_marka = islemci_marka_element.find("a")
                if a_islemci_marka:
                    islemci_marka = a_islemci_marka.text.strip()
                else:
                    islemci_marka = None
            else:
                islemci_marka = None

            islemci_seri_element = detay_soup.find("li", attrs={"id": "id1369"})
            if islemci_seri_element:
                a_islemci_seri = islemci_seri_element.find("a")
                if a_islemci_seri:
                    islemci_seri = a_islemci_seri.text.strip()
                else:
                    islemci_seri = None
            else:
                islemci_seri = None

            ram_element = detay_soup.find("li", attrs={"id": "id1027"})
            if ram_element:
                a_ram = ram_element.find("a")
                if a_ram:
                    ram = a_ram.text.strip()
                else:
                    ram = None
            else:
                ram = None

            ssd_exist_element = detay_soup.find("li", attrs={"id": "id1421"})
            if ssd_exist_element:
                a_ssd_exist = ssd_exist_element.find("a")
                if a_ssd_exist:
                    ssd_exist = a_ssd_exist.text.strip()
                else:
                    ssd_exist = None
            else:
                ssd_exist = None

            ssd_byt_element = detay_soup.find("li", attrs={"id": "id1422"})
            if ssd_byt_element:
                a_ssd_byt = ssd_byt_element.find("a")
                if a_ssd_byt:
                    ssd_byt = a_ssd_byt.text.strip()
                else:
                    ssd_byt = None
            else:
                ssd_byt = None

            gpu_exist_element = detay_soup.find("li", attrs={"id": "id1032"})
            if gpu_exist_element:
                a_gpu_exist = gpu_exist_element.find("a")
                if a_gpu_exist:
                    gpu_exist = a_gpu_exist.text.strip()
                else:
                    gpu_exist = None
            else:
                gpu_exist = None

            gpu_model_element = detay_soup.find("li", attrs={"id": "id1399"})
            if gpu_model_element:
                a_gpu_model = gpu_model_element.find("a")
                if a_gpu_model:
                    gpu_model = a_gpu_model.text.strip()
                else:
                    gpu_model = None
            else:
                gpu_model = None

            cozunurluk_element = detay_soup.find("li", attrs={"id": "id1036"})
            if cozunurluk_element:
                a_cozunurluk = cozunurluk_element.find("a")
                if a_cozunurluk:
                    cozunurluk = a_cozunurluk.text.strip()
                else:
                    cozunurluk = None
            else:
                cozunurluk = None

            pil_gucu_element = detay_soup.find("li", attrs={"id": "id1430"})
            if pil_gucu_element:
                a_pil_gucu = pil_gucu_element.find("a")
                if a_pil_gucu:
                    pil_gucu = a_pil_gucu.text.strip()
                else:
                    pil_gucu = None
            else:
                pil_gucu = None

            usb3_exist_element = detay_soup.find("li", attrs={"id": "id1435"})
            if usb3_exist_element:
                a_usb3_exist = usb3_exist_element.find("a")
                if a_usb3_exist:
                    usb3_exist = a_usb3_exist.text.strip()
                else:
                    usb3_exist = None
            else:
                usb3_exist = None

            usb3_adet_element = detay_soup.find("li", attrs={"id": "id1436"})
            if usb3_adet_element:
                span_usb3_adet = usb3_adet_element.find("span")
                if span_usb3_adet:
                    usb3_adet = span_usb3_adet.text.strip()
                else:
                    usb3_adet = None
            else:
                usb3_adet = None

            usbC_exist_element = detay_soup.find("li", attrs={"id": "id1632"})
            if usbC_exist_element:
                usbC_exist = usbC_exist_element.find("a")
                if usbC_exist:
                    usbC_exist = usbC_exist.text.strip()
                else:
                    usbC_exist = None
            else:
                usbC_exist = None

            usbC_adet_element = detay_soup.find("li", attrs={"id": "id1633"})
            if usbC_adet_element:
                usbC_adet = usbC_adet_element.find("span")
                if usbC_adet:
                    usbC_adet = usbC_adet.text.strip()
                else:
                    usbC_adet = None
            else:
                usbC_adet = None
         
            pil_hucre_element = detay_soup.find("li", attrs={"id": "id1774"})
            if pil_hucre_element:
                pil_hucre = pil_hucre_element.find("span")
                if pil_hucre:
                    pil_hucre = pil_hucre.text.strip()
                else:
                    pil_hucre = None
            else:
                pil_hucre = None

            ekrn_hız_element = detay_soup.find("li", attrs={"id": "id6294"})
            if ekrn_hız_element:
                ekrn_hız = ekrn_hız_element.find("a")
                if ekrn_hız:
                    ekrn_hız = ekrn_hız.text.strip()
                else:
                    ekrn_hız = None
            else:
                ekrn_hız = None

            hdmi_exist_element = detay_soup.find("li", attrs={"id": "id1044"})
            if hdmi_exist_element:
                hdmi_exist = hdmi_exist_element.find("a")
                if hdmi_exist:
                    hdmi_exist = hdmi_exist.text.strip()
                else:
                    hdmi_exist = None
            else:
                hdmi_exist = None

            #her biri listeye eklendi    
            urunler_liste = [
                urun_fiyat,urun_adi, urun_amaci, isletim_sis, ekrn_byt, cozunurluk, ekrn_hız,
                islemci_marka, islemci_seri, ram, ssd_exist, ssd_byt, gpu_exist, gpu_model,
                usb3_exist, usb3_adet, usbC_exist, usbC_adet, hdmi_exist,
                pil_gucu, pil_hucre
            ]
            #bir ürüne ait tüm bilgiler sırayla eklendi
            veri_listesi.append(urunler_liste)
    print(sayfa)
    sayfa+=1#linkteki sayfanın değişmesi için değer arttırıldı
df = pd.DataFrame(veri_listesi, columns=sutunlar)#dataframe e dönüştürüldü
print("Veri toplandı")
```
